```python
filenames = []
```

```python
# https://stackoverflow.com/questions/16465399/need-the-path-for-particular-files-using-os-walk
# esse aqui está funcionando OK, dá um append em filenames e depois na próxima célula joga em names.txt
for root, dirs, files in os.walk(".", topdown=False):
    for name in files:
        filenames.append( os.path.join(root,name) )
```

```python
filenames
```

```python
# https://stackoverflow.com/questions/899103/writing-a-list-to-a-file-with-python
# aqui é oficial também, guarda direitinho
with open('names.txt', 'w') as f:
    for item in filenames:
        f.write("%s\n" % item )
```

Agora vamos fazer a questão de salvar o nome do ticker (que coincidentemente é o nome do diretório) e o path

```python
filenames = []
```

```python
for root, dirs, files in os.walk(".", topdown=False):
    for name in files:
        filenames.append( (root,os.path.join(root,name)) )
```

```python
filenames
```

```python
with open('names.txt', 'w') as f:
    for item in filenames:
        f.write("%s,%s\n" % (item[0],item[1]) )
```
