#### Nome das colunas do DataFrame de Análise dos Ativos Dias

Coluna | Descrição 
----------|----------
name | Ticker do Ativo
date | Dia do Ativo
gap | Gap clássico - open de hoje em relação ao close de ontem
freefloat | FreeFloat do Ativo
price_open | Preço de abertura no core
range_core | Diferença entre high do core e low do core
is_successful | True se o close do core for menor do que o open do core. False caso contrário.
open_to_close_pct | Variação em % do open do core até o close do core
spike | High do candle onde ocorreu máxima do dia
spike_minutes | minutos passados desde 09:31 até o candle onde ocorreu a máxima do dia
low_after_high | the low of the lowest candle after the highest candle
low_after_high_minutes | minutos passados desde 09:31 até o candle onde ocorreu o low depois da máxima do dia
after_pullback_pct | variação do low(depois da máxima do dia) até o close do core. Unidade em %.
volume_pre | volume apenas do pre market
volume_core | volume apenas do core market
volume_maxima | volume apenas do core até a máxima do dia
open_pre | Open do pre market
high_pre | High do pre market
low_pre | Low do pre market
close_pre | Close do pre market
open_core | Open do core market
high_core | High do core market
low_core | Low do core market
close_core | Close do core market
